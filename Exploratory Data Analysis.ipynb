```python
plt.figure(figsize=(5,5))

ax = sns.barplot(x='subject', y='G3', hue='sex', data=data_df)

ax.set_xticklabels(('Mathematics', 'Portuguese'))
plt.xlabel('Subject')
plt.ylabel('Final Grade - G3')
plt.title('Comparison of grades between men and women by subject')
```

The above graph comapares the  grade G3 of Mathematics and Portuguese Subjects for both men and women . 
We can observe from the graph that for Mathematics Men have high grades and for Portuguese subject Women have High grades

```python
sns.boxplot(x='studytime', y='G3', data=mat_df)

plt.xlabel('Study Time')
plt.ylabel('Final Grade - G3')
plt.title('Final Grades by Study Time For Mathematics Subject')
```

```python
b = sns.swarmplot(x='G1', y='G3', data=mat_df)
b.axes.set_title('G1 vs G3 For Mathematics Subject', fontsize = 30)
b.set_xlabel('G1', fontsize = 20)
b.set_ylabel('G3', fontsize = 20)
plt.show()
```

The graph shows that the students scored high G1 has high G3 .

```python
b = sns.swarmplot(x='G1', y='G3', data=mat_df)
b.axes.set_title('G1 vs G3 For Portuguese Subject', fontsize = 30)
b.set_xlabel('G1', fontsize = 20)
b.set_ylabel('G3', fontsize = 20)
plt.show()
```

```python
b = sns.swarmplot(x='G2', y='G3', data=mat_df)
b.axes.set_title('G2 vs G3 For Mathematics Subject', fontsize = 30)
b.set_xlabel('G2', fontsize = 20)
b.set_ylabel('Final Grade', fontsize = 20)
plt.show()
```

We can observe from the plot that both G2 and G3 are directly proportional to each other.

```python
b = sns.swarmplot(x='G2', y='G3', data=por_df)
b.axes.set_title('G2 vs G3 For Portuguese Subject', fontsize = 30)
b.set_xlabel('G2', fontsize = 20)
b.set_ylabel('Final Grade', fontsize = 20)
plt.show()
```

The above graph shows that higher G2 students has higher G3 and lower G2 students has lower G2

```python
corr = mat_df.corr()
plt.figure(figsize=(10,5))
sns.heatmap(corr, cmap = 'coolwarm', annot = True)
plt.title('Correlation Matrix Of Mathematics Subject', fontsize=10);
```

From the above correlation matrix we can observe that there is strong correlation between G1,G2 and G3 for Mathematics Subject

```python
corr = por_df.corr()
plt.figure(figsize=(10,5))
sns.heatmap(corr, cmap = 'coolwarm', annot = True)
plt.title('Correlation Matrix Of Portuguese Subject', fontsize=10);
```

From the above correlation matrix we can observe that there is strong correlation between G1,G2 and G3 for Portuguese Subject
